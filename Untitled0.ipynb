```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
pd.options.display.mpl_style = 'default'
matplotlib.rcParams['figure.figsize'] = (10.0, 8.0)
import seaborn as sns
with open("per36.csv", "r") as infile:
    df = pd.read_csv(infile)
```

```python
# filter out less than 50 minutes played
df2 = df[df.MP > 50]
df2
```

```python
with open("salary.csv", "r") as infile:
    df3 = pd.read_csv(infile)
df3.head()
```

```python
# merge the two together to bring in salary
df4 = pd.merge(df2, df3[["Player", "2015-16"]], on="Player")
```

```python
import math
# just use centers and power forwards
df_final = df4[(df4["Pos"] == "C") | (df4["Pos"]=="PF")]
df_final.head()

# Normalize salary on a 1 - 5 scale
df_final['sal_norm'] = 1+100*(df_final['2015-16'] - df_final['2015-16'].min()) / (df_final['2015-16'].max() - df_final['2015-16'].min())
df_final.head()
```

```python
import math
df_final.plot(kind="scatter",x="TRB",y="PTS",s=df_final["sal_norm"])
```

```python
# write to a csv
df_output = df_final[["Player","Pos","TRB","PTS","2015-16","sal_norm"]]
with open("df_output.csv","wb") as outfile:
    df_output.to_csv(outfile)
```
